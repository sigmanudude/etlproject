```python
# Dependencies
import requests
# Dependencies and Setup
from pandas.io.json import json_normalize
import pandas as pd
import numpy as np
```

# import csv tmdb_5000_movies

```python
# File to Load (Remember to Change These)
file_to_load = "Desktop/classwork/etlproject/data/tmdb_5000_movies.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)

df.head()
```

```python
df1 = df[['title','release_date']]
df1.head()
```

```python
movies_df = df1[df.release_date > '1999-12-31'].sort_values(by='release_date', ascending=True)
```

```python
movies_df.count()
```

```python
movies_df['title'].duplicated().sum()
```

# omdb api

```python
url = "http://www.omdbapi.com/?apikey=2e886f7a&t="

title_movies = movies_df['title'].tolist()

responses = [];

for movie in title_movies:
    movie_data = requests.get(url + movie).json()
    responses.append(movie_data)
```

```python
responses_df = json_normalize(responses)
responses_df
```

```python
omdb_df = responses_df[['Title','Year','Genre', 'Awards', 'Metascore', 'BoxOffice']]

omdb_df.sort_values(by='Year', ascending=True)
```

```python
omdb_df.count()
```

```python
new_movies = omdb_df.dropna(how="any")

new_movies.count()
```

```python
new_movies["Year"] = new_movies.Year.apply(int)
```

```python
new_movies.count()
```

```python
cool_movies = new_movies[new_movies.Year > 1999].sort_values(by='Year', ascending=True)
```

```python
cool_movies.count()
```
