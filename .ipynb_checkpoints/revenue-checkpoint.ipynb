```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# File to Load (Remember to Change These)
file_to_load = "~/Desktop/Movies_Merged.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)

df.head()
```

```python
# Filter the DataFrame down only to those columns to chart
title_rev = df[["Title","revenue"]]

# Set the index to be "State" so they will be used as labels
#title_rev = df.set_index("revenue")

title_rev.head()
```

```python
revenue_df = title_rev[title_rev.revenue != 0].sort_values(by='revenue', ascending=False)

revenue_df['revenue'] = revenue_df[['revenue']] / 1000000

revenue_df = revenue_df.set_index("Title")

top_15 = revenue_df.head(15)
```

```python
top_15
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
top_15.plot(kind="bar", figsize=(9,7))

# Set a title for the chart
plt.title("Top 10 Highest Grossing Films")
plt.xlabel("Title")
plt.ylabel("Revenue / 1000000")
#plt.savefig("~/Desktop/classwork/etlproject/static/images/top5.png")
plt.tight_layout()
plt.savefig("C:/Users/justi/Desktop/classwork/etlproject/static/Top15grossing.png")
plt.show()
```

```python
least_revenue_df = title_rev[title_rev.revenue != 0].sort_values(by='revenue', ascending=True)

least_revenue_df['revenue'] = least_revenue_df[['revenue']] / 1000000

least_revenue_df = least_revenue_df.set_index("Title")

flops = least_revenue_df.head(15)

flops
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
flops.plot(kind="barh", figsize=(9,7))

# Set a title for the chart
plt.title("15 Least Grossing")
plt.xlabel("Title", fontsize=12)
plt.ylabel("Revenue / 1000000")
plt.tight_layout()
plt.savefig("C:/Users/justi/Desktop/classwork/etlproject/static/15flops.png")
plt.show()

```
