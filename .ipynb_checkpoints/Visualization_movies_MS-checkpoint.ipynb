```python
# Import all required dependencies
import csv
import pandas as pd
import numpy as np
import os
from datetime import datetime
import calendar

import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns

from scipy.stats import linregress

import scipy.stats as stats

#from spread import variance, standard_deviation, zipped_z_scores

import warnings
warnings.filterwarnings('ignore')

    
```

```python
all_mov_data= pd.read_csv("Movies_Merged.csv", encoding = 'utf8')

all_mov_data.head()
```

```python
all_mov_data.count()
```

```python
# rename col 1 to ID
all_mov_data.rename(columns = {'Unnamed: 0': 'movie_id'}, inplace = True)
all_mov_data.columns
```

```python
all_mov_data.describe()
```

```python
all_mov_data[['budget_fmt']] = all_mov_data[['budget']] / 1000000
```

```python
all_mov_data[['revenue_fmt']] = all_mov_data[['revenue']] / 1000000
```

```python
# convert all TMDB vote average to a 100 scale so that comparison is an even field with Metascore
all_mov_data[['vote_average']] = all_mov_data[['vote_average']]*10
```

```python
all_mov_data[['Title','vote_average','Metascore']].head(10)
```

```python
fig, ax = plt.subplots(figsize = (9,7))
sns.distplot(all_mov_data.vote_average, bins = 15, ax=ax, color = '#0098BF' )
sns.distplot(all_mov_data.Metascore, bins = 15, ax=ax, color = '#FF9900')

ax.set_title("Rating of Movies - Metascore critics vs. TMDB Audience Rating", fontsize = 15)
ax.set_xlabel("Metascore critics vs. TMDB audience rating", fontsize = 13)
ax.set_ylabel("Frequency", fontsize = 13)

plt.tight_layout()
plt.savefig("static/images/metascore_tmdb_rating_compare.png")
plt.show()
```

```python
mov_status_budget = all_mov_data[['status','budget_fmt']].groupby('status').agg(np.mean)

mov_status_budget
```

```python
mov_bgt_cnt = all_mov_data[['Year','budget_fmt']].groupby('Year').agg(['sum','count'])
mov_bgt_cnt.reset_index(inplace = True)
mov_bgt_cnt.set_axis(1, range(len(mov_bgt_cnt.columns)))

mov_bgt_cnt.rename(columns = {0:'Year',1 : 'Budget', 2:'MovieCount'}, inplace = True)
mov_bgt_cnt
```

```python
# Study Budget spents and Movies produced across Years

fig, ax1 = plt.subplots(figsize = (9,7))

color = '#0098BF'
ax1.set_xlabel('Year', size=15)
ax1.set_ylabel('Total Budget (in USD millions)', color=color, size=15)
ax1.bar(mov_bgt_cnt.Year, mov_bgt_cnt.Budget, color=color, linewidth = 5, alpha = 0.5)
ax1.tick_params(axis='y', labelcolor=color)
# ax1.set_ylim(16, 18, 1)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = '#FF9900'
ax2.set_ylabel('# of Movies Made', color=color,size=15)  # we already handled the x-label with ax1
ax2.plot(mov_bgt_cnt.Year, mov_bgt_cnt.MovieCount, color=color, linewidth = 5)
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Number of Movies made against Budget spent over Years', size=20)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig('static/images/budget_movcnt_yrs.png')

plt.show()
```

```python
all_mov_data['Genre_Type1'] = all_mov_data.Genre.apply(lambda s : s.split(",")[0])
```

```python
MaxBgt_flops = all_mov_data.loc[(all_mov_data.budget_fmt > 0) & (all_mov_data.revenue_fmt > 0)]
MaxBgt_flops = MaxBgt_flops.sort_values(['budget_fmt','revenue_fmt'],
                                        ascending = [False,True])
```

```python
MaxBgt_flops = MaxBgt_flops[['Title','budget_fmt','revenue_fmt']].tail(15)
MaxBgt_flops.dtypes
```

```python
fig, ax = plt.subplots(figsize = (9,7))

sns.barplot(x = 'revenue_fmt', y = 'Title', data = MaxBgt_flops, label = "revenue (in millions)", color = "#FF9900", alpha = 0.5)
sns.barplot(x = 'budget_fmt', y = 'Title', data = MaxBgt_flops, label = "Budget (in millions)", color = "#0098BF")
ax.legend(ncol=1, loc='best', frameon=True)
# sns.despine(left=True, bottom=True)

ax.set_title("Top 15 movies - Revenue vs Budget", fontsize = 15)
ax.set_xlabel("Revenue (in millions) vs Budget (in millions)", fontsize = 12)
ax.set_ylabel("")

plt.tight_layout()
plt.savefig("static/images/Top15.png")
plt.show()
```
