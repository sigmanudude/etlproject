```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# File to Load (Remember to Change These)
file_to_load = "~/Desktop/Movies_Merged.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)

df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, df["revenue"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, df["Title"], rotation="vertical")
```

```python
# Filter the DataFrame down only to those columns to chart
df = df[["Title","revenue"]]

# Set the index to be "State" so they will be used as labels
title_rev = df.set_index("revenue")

title_rev.head()
```

```python
revenue_df = title_rev[title_rev.revenue != 0].sort_values(by='revenue', ascending=False)

divideby = revenue_df[['revenue']] / 1000000

top_5 = divideby.head(10)
```

```python
divideby.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
top_5.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("Top 10 Highest Grossing Films")
plt.xlabel("Title")
plt.ylabel("Revenue / 1000000")
plt.savefig("~/Desktop/classwork/etlproject/")
plt.show()
plt.tight_layout()
```

```python
least = title_rev[title_rev.revenue != 0].sort_values(by='revenue', ascending=True)
flops = least.head(10)
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
flops.plot(kind="bar", figsize=(20,3))

# Set a title for the chart
plt.title("10 Flops")
plt.xlabel("Title")
plt.ylabel("Revenue / 1000000")
plt.show()
plt.tight_layout()
```
